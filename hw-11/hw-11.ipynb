```python
fraud_pct = df['Class'].value_counts(normalize=True)[1] * 100
print('Процент аномальных транзакций: {:.2f}%'.format(fraud_pct))
```

```python
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

contamination = fraud_pct / 100
```

```python
X_train, X_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
models = [('Isolation Forest', IsolationForest(n_estimators=100, contamination=fraud_pct/100)),
          ('OneClassSVM', OneClassSVM(kernel='rbf', nu=1-contamination, gamma=0.1))]

for name, model in models:
    model.fit(X_train.drop(['Class'], axis=1))

    y_train_pred = model.predict(X_train.drop(['Class'], axis=1))
    y_test_pred = model.predict(X_test.drop(['Class'], axis=1))

    print(name)
    print('Train')
    print(confusion_matrix(X_train['Class'], y_train_pred))
    print(classification_report(X_train['Class'], y_train_pred))
    print('Test')
    print(confusion_matrix(X_test['Class'], y_test_pred))
    print(classification_report(X_test['Class'], y_test_pred))
```

```python
from sklearn.manifold import TSNE

y = df['Class']

tsne = TSNE(random_state=42)

X_tsne = tsne.fit_transform(df.drop(['Class'], axis=1))

plt.scatter(X_tsne[y==0, 0], X_tsne[y==0, 1], label="Normal", alpha=0.2)
plt.scatter(X_tsne[y==1, 0], X_tsne[y==1, 1], label="Fraud", alpha=0.2, color="red")
plt.legend()
plt.show()
```
