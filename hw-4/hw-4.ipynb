## Обучение модели 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data.drop(['diagnosis'], axis=1), data['diagnosis'], test_size=0.30, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#в данном случае нормализация нужна для того, чтобы привести признаки к одному масштабу и повысить эффективность и точность при обучении модели
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred, pos_label='M')
```

```python
recall_score(y_test, y_pred, pos_label='M')
```

```python
f1_score(y_test, y_pred, pos_label='M')
```

```python
from sklearn.metrics import roc_curve
import numpy as np
```

## Построение ROC-кривой

```python
y_pred_proba = knn.predict_proba(X_test_scaled)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba[:, 1], pos_label='M')
```

```python
plt.plot(fpr, tpr)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
area_under_curve = roc_auc_score(y_test, y_pred_proba[:, 1])
print(area_under_curve)
```

## Подбор гиперпараметров

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_neighbors':[1, 2, 5, 10, 15, 20, 30, 40, 50, 75, 100]}
knn = KNeighborsClassifier()
clf = GridSearchCV(knn, parameters,cv = 5, scoring = 'accuracy')
clf.fit(X_train, y_train)
print('Best params: ', clf.best_params_)
print('Best accuracy: ', clf.best_score_)
```
